```python
def first_element(l):
    return l[0]
first_element_vectorized = np.vectorize(first_element)
```

```python
def second_element(l):
    return l[1]
second_element_vectorized = np.vectorize(second_element)
```

```python
def third_element(l):
    return l[2]
third_element_vectorized = np.vectorize(third_element)
```

```python
separated_code = second_element_vectorized(separated_title)
separated_code = np.char.split(separated_code, ")")
separated_code
```

```python
separated_title = first_element_vectorized(separated_title)
separated_title
```

```python
# has a few errors that i will fix manually
separated_code = first_element_vectorized(separated_code)
separated_code
```

```python
def link_generator(s):
    generic_link = f"https://didattica.unibocconi.it/ts/tsn_anteprima.php?cod_ins={s}&anno=2024&ric_cdl=TR10&IdPag=7399"
    return generic_link
vectorized_link_generator = np.vectorize(link_generator)
```

```python
links = vectorized_link_generator(separated_code)
```

```python
columns = np.array([separated_title, separated_code, links])
```

```python
spreadsheet_df = pd.DataFrame({"Title": columns[0], "Code": columns[1], "Link": columns[2]})
spreadsheet_df
```

```python
spreadsheet_df.to_csv("Elective classes with BEMACS links.csv", sep=",", index=False, encoding="utf-8")
```
